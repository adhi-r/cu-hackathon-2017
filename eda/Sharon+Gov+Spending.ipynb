```python
df_list = []
for filename in sorted(dat_2000_2012):
    data = pd.read_csv(filename, usecols=header_index)
 #   data.columns = map(str.lower, data.columns) #remove non-unique header data
   # print(filename, data.columns)
  #  data['filename']= str.replace(filename,'/data/govt2000_2012/GovernmentSpendingContracts-','')
    df_list.append(data) # modify this to limit the number of rows to extract
    
full_df = pd.concat(df_list)
```

```python
#full_df['year'] = pd.to_numeric(full_df['fiscal_year'].str[:4])#.to_numeric()
#bad_years = full_df[full_df['year'] < 1999 ]
full_df = full_df[full_df['fiscal_year'] > 1999]
full_df = full_df[full_df['fiscal_year'] < 2017]
full_df = full_df[full_df['base_and_all_options_value'] < 10000000]
full_df = full_df[full_df['base_and_all_options_value'] >= 0] 
idc = full_df['contract_action_type'] == "IDC Indefinite Delivery Contract"
full_df = full_df[~idc]

#clean_data.groupby('year').count()
#print(years)
#full_df.head()
```
