```python
import pandas as pd
import glob
dat_2000_2012 = glob.glob("/data/govt2000_2012/*.csv")
df_list = []
header_index = []

#grab location of columns of interest
for entries in ["unique_transaction_id", "women_owned_flag"]:
    header_index.append(header_list.index(entries))
    
for filename in sorted(dat_2000_2012):
    data = pd.read_csv(filename, usecols=header_index)
 #   data.columns = map(str.lower, data.columns) #remove non-unique header data
   # print(filename, data.columns)
  #  data['filename']= str.replace(filename,'/data/govt2000_2012/GovernmentSpendingContracts-','')
    df_list.append(data) # modify this to limit the number of rows to extract
    
full_df = pd.concat(df_list)
```
